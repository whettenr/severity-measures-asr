# Comparing

### Cos distance

```python
x = np.arange(0, 50, 1)

plt.plot(x, history_base['train_cos_dis'], label='Base Model')
plt.plot(x, history_cos['train_cos_dis'], label='CTC + Cos Model')
plt.plot(x, history_cos2['train_cos_dis'], label='CTC + Cos^2 Model')
# plt.plot(x, y2, '-.')

plt.legend(loc='upper right')
plt.xlabel("Epoch")
plt.ylabel("Cos Dis")
plt.title('Model Comparison: Training Cos Dis')
plt.show()
```

```python
x = np.arange(0, 50, 1)

plt.plot(x, history_base['val_cos_dis'], label='Base Model')
plt.plot(x, history_cos['val_cos_dis'], label='CTC + Cos Model')
plt.plot(x, history_cos2['val_cos_dis'], label='CTC + Cos^2 Model')
# plt.plot(x, y2, '-.')

plt.legend(loc='upper right')
plt.xlabel("Epoch")
plt.ylabel("Cos Dis")
plt.title('Model Comparison: Validation Cos Dis')
plt.show()
```
